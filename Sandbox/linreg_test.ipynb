```python
from linear_regression.linear_regression import LinearRegression
from my_micrograd.graph import draw_dot
import matplotlib.pyplot as plt
```

```python
xs = [
    [0, 1, 0.5, 0.3],
    [1, 0.5, 0.5, 0.2],
    [0, 0.1, 1, 0],
    [0.5, 0.7, 1, 0.7],
    [1, 0, 0.1, 0.8]
]
ys = [0, 1, 1, 0, 0]

reg = LinearRegression(4)
```

```python
reg.predict(xs)
```

```python
loss = reg._mean_squared_error(xs, ys)
draw_dot(loss)
```

```python
history = reg.fit(xs, ys, 1000, learning_rate=0.01)
```

```python
plt.plot([h.data for h in history])
plt.show()
```
